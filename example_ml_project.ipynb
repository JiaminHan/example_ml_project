```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#handling null values
df.isnull().sum()
```

```python
#calculate missing percentage for each feature
mis_count=df.isnull().sum()
mis_percentage=df.isnull().sum()/len(df)
mis_table=pd.concat([mis_count,mis_percentage],axis=1)
mis_table.rename(columns={0:'mis_count',1:'mis_percentage'},inplace=True)
mis_table.sort_values(by='mis_percentage',ascending=False)
```

```python
#drop columns have missing percentage larger than 60%
drop_columns=list(mis_table[mis_table['mis_percentage']>0.6].index)
drop_columns
```

```python
df.drop(columns=drop_columns,inplace=True)
```

```python
mis_table[(mis_table['mis_percentage']>0)&(mis_table['mis_percentage']<=.6)].index
```

```python
#fillna
fillna_list=list(mis_table[(mis_table['mis_percentage']>0)&(mis_table['mis_percentage']<=.6)].index)
for i in fillna_list:
    df[i].fillna(df[i].median(),inplace=True)
```

```python
df.isnull().sum().sum()
```

```python
# handling duplicates
df.duplicated().sum() #no duplicates identified
```

```python
# visualize target variable
plt.hist(df['X42'],bins=50);
```

```python
# remove outliers
upper_bound=df['X42'].mean()+df['X42'].std()*3
lower_bound=df['X42'].mean()-df['X42'].std()*3
df=df[(df['X42']<upper_bound)&(df['X42']>lower_bound)]
plt.hist(df['X42'],bins=50);
```

```python
# build model and evaluate
#data preprocess
obs=df.copy()
cat_list=['X19','X20','X21','X22','X24','X26','X28','X41']
for i in cat_list:
    obs[i]=obs[i].astype(str)
```

```python
obs=pd.get_dummies(obs,columns=cat_list)
```

```python
obs.columns
```

```python
from sklearn.model_selection import train_test_split
X=obs.drop(columns='X42')
y=obs['X42']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.columns
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn_pandas import DataFrameMapper
mapper = DataFrameMapper([ 
    (['X1', 'X2', 'X3', 'X4', 'X6', 'X7', 'X8', 'X9', 'X10', 'X11', 'X12',
       'X13', 'X14', 'X15', 'X16', 'X17', 'X18', 'X23', 'X25', 'X27', 'X29',
       'X30', 'X31', 'X32', 'X33', 'X34', 'X35', 'X39', 'X40'], [StandardScaler()]),
        (['X19_0','X19_1', 'X19_2', 'X19_3', 'X20_0', 'X20_1', 'X20_2', 'X21_0', 'X21_1',
       'X21_2', 'X21_3', 'X22_0', 'X22_1', 'X22_2', 'X24_0', 'X24_1', 'X24_2',
       'X24_3', 'X26_0', 'X26_1', 'X26_2', 'X26_3', 'X28_0', 'X28_1', 'X28_2',
       'X28_3', 'X28_4', 'X41_2006', 'X41_2007', 'X41_2008', 'X41_2009',
       'X41_2010'], None)])

X_train=mapper.fit_transform(X_train)
X_test=mapper.transform(X_test)
```

```python
#model selection
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error
```

```python
def fit_and_evaluation(model):
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    mae=mean_absolute_error(y_test,y_pred)
    return mae
```

```python
lasso=Lasso()
fit_and_evaluation(lasso)
```

```python
rf=RandomForestRegressor()
fit_and_evaluation(rf)
```

```python
gb=GradientBoostingRegressor()
fit_and_evaluation(gb)
```

```python
plt.scatter(gb.predict(X_test),y_test,alpha=0.5);
```

```python
#optimization
from sklearn.model_selection import RandomizedSearchCV
param_dist = {"max_depth": np.arange(1,15,3),
              "max_features":['sqrt','log2'],
              'n_estimators':np.arange(10,250,50),
              'learning_rate':np.logspace(-2,0,4)}

# run randomized search
n_iter_search = 20
random_search = RandomizedSearchCV(gb, param_distributions=param_dist,
                                   n_iter=n_iter_search, cv=5)

random_search.fit(X_train, y_train)
```

```python
final_model=random_search.best_estimator_
fit_and_evaluation(final_model)
```

```python
plt.scatter(y_test,final_model.predict(X_test),alpha=0.2);
```
